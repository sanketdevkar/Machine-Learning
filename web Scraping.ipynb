```python
import pandas as pd     
import  requests
from bs4 import BeautifulSoup


name=[]
rating=[]
rating_count=[]
review_count=[]
info=[]
final_df=[]
for page in range(1,3):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage=requests.get(f'https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_6_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_6_0_na_na_na&as-pos=6&as-type=TRENDING&suggestionId=mobiles&requestId=7487beee-13e1-4b87-a1f9-367d5e46e0c6&page={page}',headers=headers).text

    soup =BeautifulSoup(webpage,'lxml')
    count=soup.find_all('span',class_="PvbNMB")
    uls = soup.find_all('ul', class_="HwRTzP")
    for i in range(24):
        name.append(soup.find_all('div',class_="RG5Slk")[i].text)
        rating.append(soup.find_all('div',class_="MKiFS6")[i].text)
        rating_count.append(count[i].text.split()[0])
        review_count.append(soup.find_all('span',class_="PvbNMB")[i].text.replace('\xa0',' ').split('&')[1].split()[0])
        items = uls[i].find_all('li', class_="DTBslk")
        phone_info = [item.text for item in items]
        info.append(phone_info)

    
d={'name':name,'rating':rating,'rating_count':rating_count,'review_count':review_count,'info':info}

df=pd.DataFrame(d)
df.to_csv('flipkart_mobiles.csv',index=False)


pd.read_csv('flipkart_mobiles.csv')
```
